# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here
salaries.describe()
```

```python
import math
from scipy import stats

sample_mean = 32.788558
sample_size = 8022
sample_std_dev = 12.112573
expected_mean = 30

# H0: Null hypothesis
# The average hourly salary does not differ from $30/hr.
# The true population mean is exactly $30/hr.

# Standard error
sem = sample_std_dev / math.sqrt(sample_size)

# Calculate the test statistic (how far the sample mean is from the hypothesized mean)
t_statistic = (sample_mean - expected_mean) / sem

# Two-sided p-value (we are testing if the true mean is either lower OR higher than $30/hr)
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=sample_size - 1))

# Significance level
alpha = 0.05

# Output the result
print(f"Test statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Conclusion: The null hypothesis is rejected.")
    print("The average hourly salary differs significantly from $30/hr.")
    print("It is unlikely that the observed sample mean ($32.79) occurred by chance if the true mean were $30/hr.")
else:
    print("Conclusion: We fail to reject the null hypothesis.")
    print("There is no statistically significant evidence that the average hourly salary differs from $30/hr.")
```

# Challenge 3 - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval.
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers.

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here

sample_mean = 32.788558
sample_size = 8022
sample_std_dev = 12.112573

# Standard error
sem = sample_std_dev / math.sqrt(sample_size)

# 95%-confidence-intervall
confidence_level = 0.95
df = sample_size - 1  # degree of freedom
confidence_interval = stats.t.interval(confidence_level, df, loc=sample_mean, scale=sem)

# outcome of confidence intervall
print(f"The 95%-confidence intervall of the average hourly salary is: {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f} $/h")
```

# Challenge 4 - Hypothesis Tests of Proportions

Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction.

You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).

In the cell below, use the `proportions_ztest` function from `statsmodels` to perform a hypothesis test that will determine whether the number of hourly workers in the City of Chicago is significantly different from 25% at the 95% confidence level.

```python
from statsmodels.stats.proportion import proportions_ztest
import math
from scipy.stats import norm

# Data
n = 25161
successes = 8022  # hourly workers
expected_prop = 0.25

# Hypothesis test
stat, p_value = proportions_ztest(count=successes, nobs=n, value=expected_prop)

# Sample proportion
p_hat = successes / n

# Confidence level
confidence_level = 0.95
z_critical = norm.ppf((1 + confidence_level) / 2)

# Standard error
se = math.sqrt((p_hat * (1 - p_hat)) / n)

# Confidence interval
ci_lower = p_hat - z_critical * se
ci_upper = p_hat + z_critical * se

# Output
print(f"z-statistic: {stat:.2f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("→ The null hypothesis is rejected: The proportion of hourly workers is significantly different from 25%.")
else:
    print("→ The null hypothesis is not rejected: The proportion of hourly workers is not significantly different from 25%.")

print(f"95% confidence interval for the proportion of hourly workers: {ci_lower:.3f} to {ci_upper:.3f}")
```
